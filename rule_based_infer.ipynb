```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import make_scorer
from joblib import dump, load
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer, IterativeImputer, KNNImputer
from sklearn.base import RegressorMixin, BaseEstimator
```

```python
def rull_based_impute(df: pd.DataFrame):
    df_inferred = df.copy()
    for idx, row in df_inferred.iterrows():
        if np.isnan(row.AGE) and not isinstance(df_inferred.loc[idx,"VISCODE"], float):
            diff = df_inferred.loc[idx, ["VISCODE", "AGE"]].apply(lambda x: x[1]-x[0]/12, axis=1).mean()
            df_inferred.loc[idx, "AGE"] = df_inferred.loc[idx, "VISCODE"]/12 + diff

        if np.isnan(row.PTGENDER_num) and not isinstance(df_inferred.loc[idx,"PTGENDER_num"], float):
            df_inferred.loc[idx, "PTGENDER_num"] = df_inferred.loc[idx, "PTGENDER_num"].mode(dropna=True)

        if np.isnan(row.PTEDUCAT) and not isinstance(df_inferred.loc[idx,"PTEDUCAT"], float):
            df_inferred.loc[idx, "PTEDUCAT"] = df_inferred.loc[idx, "PTEDUCAT"].max()
    return df_inferred
```

```python
df_dev_set = pd.read_csv("genentech-404-challenge/dev_set.csv").set_index(["RID_HASH"])
df_dev_1 = pd.read_csv("genentech-404-challenge/dev_1.csv").set_index(["RID_HASH"])
df_dev_2 = pd.read_csv("genentech-404-challenge/dev_2.csv").set_index(["RID_HASH"])
df_dev_3 = pd.read_csv("genentech-404-challenge/dev_3.csv").set_index(["RID_HASH"])

df_test_a = pd.read_csv("genentech-404-challenge/test_A.csv").set_index(["RID_HASH"])
df_test_b = pd.read_csv("genentech-404-challenge/test_B.csv").set_index(["RID_HASH"])


df_test_a_inferred = rull_based_impute(df_test_a)
df_test_b_inferred = rull_based_impute(df_test_b)
```

```python
df_test_a_inferred.to_csv("genentech-404-challenge/test_A_inferred.csv")
df_test_b_inferred.to_csv("genentech-404-challenge/test_B_inferred.csv")
```
